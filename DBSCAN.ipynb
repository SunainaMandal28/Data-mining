```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Generate sample data (two interleaving half circles)
X, y = make_moons(n_samples=300, noise=0.08, random_state=0)

# Fit DBSCAN model
db = DBSCAN(eps=0.25, min_samples=5).fit(X)
labels = db.labels_

# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', alpha=0.7)
plt.title('DBSCAN Clustering')
plt.show()
```
